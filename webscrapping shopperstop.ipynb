```python
#PROBLEM STATEMENT

              #TO COLLECT AND STORE THE DATA OF TOP WATCHES FORM THE SHOPPERSTOP WEBSITE USING WEBSCRAPPING

#IMPORTING THE REQURIED LIBRARIES

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import pandas as pd
import codecs
import re
from webdriver_manager.chrome import ChromeDriverManager
import time
```

```python
#TO OPEN THE CHROME BROWSER

driver=webdriver.Chrome(service=Service(ChromeDriverManager().install()))

#ENTERING THE URL 

val = "https://www.shoppersstop.com/men-watches/c-A1080"

#WAIT AND EC USED FOR THE TIME TO LOAD THE WEBSITE

wait = WebDriverWait(driver, 10)
driver.get(val)
get_url = driver.current_url
wait.until(EC.url_to_be(val))

#GETTING PRODUCTDETAILS

def pro_data(soup,element,classname): 
    data_str = "" 
    product_name = [] 
    product_code=[]
    brand_name = []
    price = []
    category=[]
    category_code = []
  
 #READING ALL PRODUCT DETAILS THORUGH THE ELEMENT ATTRIBUTES 
    
    for item in soup.find_all(element, class_=classname): 
        product_name.append(item.get("data-product-name"))
        product_code.append(item.get("data-product-code"))
        brand_name.append(item.get("data-product-brandname"))
        price.append(item.get("data-product-price"))
        category_code.append(item.get("data-category-code"))
        category.append(item.get("data-category-name"))
        
 # SAVING THE DATA AS DICTIONARY

    data = {'product_name'  : product_name,
            'product_code'    : product_code,
            'brand_name ':brand_name,
            'price':price,
            'category': category,
            'category_code':category_code}
    return data 

#CHECK WHETHER THE DATA IS FETCHED FROM SOURCE

if get_url == val:
    
    
#SCROLLING FOR 5 TIMES TO GET TO THE BOTTOM OF THE PAGE TO EXTRACT ALL DATA

    for i in range(0,5):
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        
    page_source = driver.page_source
    time.sleep(30) #SLEEP FOR 30 SECONDS TO ENSURE ALL THE DATA HAS BEEN LOADED
    
#GETTING THE HTML DATA OF THE WEBSITE

    soup = BeautifulSoup(page_source,features="html.parser")
    keyword="body"
    matches = soup.body.find_all(string=re.compile(keyword))
    len_match = len(matches)
    title = soup.title.text
    
#FUNCTION CALL TO GET THE PRODUCT DETAILS

    product_details = pro_data(soup,"li","pro-box") 
    
#CLOSING THE BROWSER AND TERMINATING THE DRIVER

driver.close()    
driver.quit()
```

```python
# Create DataFrame 
df = pd.DataFrame(product_details) 
df.head()


# Save the output. 
df.to_csv('SHOPERSSTOP_WATCHES.csv') 
```
